```python
import pandas as pd
import numpy as np
import geopandas as gpd
import os
import glob
import re
from functools import reduce
```

```python
RawDataPath = r"D:\CivicDataLabs_IDS-DRR\IDS-DRR_Github\HP\flood-data-ecosystem-Himachal-Pradesh\Sources\BHARATMAPS\data\RawData"
#road_length = glob.glob('*sdt_road_length.csv')
df1 = gpd.read_file(r'tehsil_SH.geojson')
df2 = gpd.read_file(r'tehsil_NH.geojson')
df3 = gpd.read_file(r'tehsil_allroads.geojson')

road_length = [df1,df2,df3]

df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['object_id'],
                                            how='outer'), road_length)

df_merged = df_merged.fillna(0)

df_merged['road_length']=df_merged['LENGTH']+df_merged['LENGTH_x']+df_merged['LENGTH_y']
df_merged['road_count']=df_merged['COUNT']+df_merged['COUNT_x']+df_merged['COUNT_y']
df_merged=df_merged.drop(['District_x','STATE_x', 'TEHSIL_x', 'objectid_x', 'dtcode11_x',
       'objectid', 'LENGTH_x', 'COUNT_x', 'geometry_x', 'District_y',
       'STATE_y', 'TEHSIL_y', 'objectid_y', 'dtcode11_y', 'LENGTH_y',
       'COUNT_y', 'geometry_y','STATE', 'dtcode11', 'LENGTH', 'COUNT', 'geometry'],axis=1)
df_merged
```

```python
BM_roads = df_merged.to_csv(r'D:\CivicDataLab_IDS-DRR\IDS-DRR_Github\HP\flood-data-ecosystem-Himachal-Pradesh\Sources\BHARATMAPS\data\variables\RoadLength\RoadLengths.csv', index=False)
```
