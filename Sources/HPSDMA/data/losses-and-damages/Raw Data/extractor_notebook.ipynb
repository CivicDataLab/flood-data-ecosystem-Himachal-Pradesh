```python
import re
from pathlib import Path
import pandas as pd
```

```python
SQL_FILE = r"D:\CivicDataLab_IDS-DRR\IDS-DRR_Github\Deployment\flood-data-ecosystem-Himachal-Pradesh\Sources\HPSDMA\data\losses-and-damages\Raw Data\losses_export_script.sql"

# Read file (handle BOM and encoding)
text = Path(SQL_FILE).read_bytes()
if text.startswith(b'\xef\xbb\xbf'):
    text = text[3:]
text = text.decode("utf-8", errors="ignore")

# Regex to match CREATE TABLE [schema].[table]
table_regex = re.compile(
    r'CREATE\s+TABLE\s+\[\s*(\w+)\s*\]\.\[\s*(\w+)\s*\]',
    re.IGNORECASE
)

print(text[:1000])
```

```python
text = Path(SQL_FILE).read_bytes()
if text.startswith(b'\xff\xfe'):   # BOM for UTF-16 LE
    text = text.decode("utf-16")   # decode properly
else:
    text = text.decode("utf-8", errors="ignore")

print(text[:1000])
```

```python

tables = set()
for match in table_regex.finditer(text):
    schema, table = match.groups()
    tables.add(f"{schema}.{table}")

print("Tables found:")
for t in sorted(tables):
    print(t)
```
