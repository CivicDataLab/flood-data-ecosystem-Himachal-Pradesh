```python
import pandas as pd
```

```python
project_root = '/home/krishna/IDS-DRR-Data-Pipeline/Hyderabad/'
```

```python
waterlogging_df_dict = pd.read_excel(project_root+'Data/Water Logging point, GHMC Limits-2023(New).xlsx',
                               sheet_name = ['MAJOR', 'MEDIUM', 'MINOR'])
```

```python
waterlogging_df_dict['MAJOR'][:30].dropna(axis = 0, how = 'all')
```

```python

# Rrepare water logging csv with following columns
## WATER LOGGING TYPE -- Keys of df_dicts
waterlogging_df = pd.DataFrame(columns = ['type', 'team', 'zone', 'location', 'landmark', 'lat', 'long'])


for TYPE in waterlogging_df_dict.keys():
    df = waterlogging_df_dict[TYPE]
    df = df.dropna(axis = 0, how = 'all')
    ZONE= ''
    for row in df.itertuples():
        ## itertuples row[0] is index
    ## ZONE -- first col that has 'ZONE' in it  -- don't change the value until new zone is identified  
        if ('zone' in str(row[1]).lower()) | (ZONE==''):
            ZONE = row[1]
            continue # Go to nxt row - nothing to retreive else from this row
        
        if 'long' in str(row[5]).lower():
            continue # Go to nxt row - nothing to retreive  from this row
            
        TEAM = str(row[2])
        LOCATION = row[3]
        LANDMARK = row[4]
        LAT = row[5]
        LONG = row[6]
        
              
        append_df = pd.DataFrame([TYPE, TEAM, ZONE, LOCATION, LANDMARK, LAT, LONG],
                                index = ['type', 'team', 'zone', 'location', 'landmark', 'lat', 'long']).T
        
        waterlogging_df = pd.concat([waterlogging_df,append_df])
```

```python
# Cleaning - 
waterlogging_df = waterlogging_df.reset_index(drop=True)
## remove duplicate header columns
index_duplicateheaders = waterlogging_df[ (waterlogging_df['lat'].str.contains('LAT').fillna(False)) ].index
waterlogging_df = waterlogging_df.drop(index_duplicateheaders)
## Fill NULLS - bfills
waterlogging_df['team'] = waterlogging_df['team'].ffill()

waterlogging_df = waterlogging_df.reset_index(drop=True)
```

```python
waterlogging_df.to_csv(project_root+'Data/Water Logging point, GHMC Limits-2023(New).csv')
```
