```python
import pandas as pd

def tender_count_by_fy(csv_path: str) -> pd.DataFrame:
    df = pd.read_csv(csv_path)
    df["Published Date"] = pd.to_datetime(
        df["Published Date"], format="%d-%b-%Y %I:%M %p", errors="coerce"
    )
    def to_fy(dt):
        if pd.isna(dt): return None
        year = dt.year if dt.month >= 4 else dt.year - 1
        return f"{year}-{str(year + 1)[2:]}"
    df["FY"] = df["Published Date"].apply(to_fy)
    return df.groupby("FY").size().reset_index(name="Count")

result = tender_count_by_fy("data/flood_tenders_all.csv")
print(result.to_string(index=False))
print(f"\nTotal: {result['Count'].sum()}")
```

generate a function that will give the annual value of tenders in the financial year 2025-2026 from /Users/saurabhlevin/Deployment/flood-data-ecosystem-Assam/Sources/TENDERS/data/assam_all_tenders.csv


```python
import pandas as pd

def annual_tender_value_fy2025_26(csv_path: str) -> tuple[float, int]:
    df = pd.read_csv(csv_path)
    df["Published Date"] = pd.to_datetime(
        df["Published Date"], format="%d-%b-%Y %I:%M %p", errors="coerce"
    )
    mask = (df["Published Date"] >= "2025-04-01") & \
           (df["Published Date"] <= "2026-03-31 23:59:59")
    fy_df = df[mask].copy()
    fy_df["_value"] = (
        fy_df["Tender Value in ₹"]
        .astype(str)
        .str.replace(",", "", regex=False)
        .pipe(pd.to_numeric, errors="coerce")
    )
    return fy_df["_value"].sum(), int(mask.sum())

total, count = annual_tender_value_fy2025_26("data/assam_all_tenders.csv")
print(f"FY 2025-26 — {count} tenders, Total Value: ₹{total:,.0f}")
```
